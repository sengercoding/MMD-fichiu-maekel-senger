**Task:** 
Consider a marriage agency in charge of pairing persons based on their similarities and
common interests. Each customer has to fill-in b questionnaires Q1, . . . , Qb containing r
questions each. Two customers are matched if they give exactly the same answers to all r
questions for at least one of these b questionnaires. Assume that two specific customers
C1 and C2 give with probability p the same answer to any of the questions in any of
the questionnaires. State formulas (and explain how you derived them) for the following
cases:

a)

**Task:**

The probability that C1 and C2 are matched.

**Solution:**

The probability that C1 and C2 are matched. 

= 1 - The probability that C1 and C2 dont match.

= 1- (Probability that C1 and C2 dont match with a single Questionaire)^b

Probability that C1 and C2 dont match with a single Questionaire 

= 1 - Probability that C1 and C2  match with a single Questionaire 

= 1 - p^r

=> The probability that C1 and C2 are matched = 1 - (1 - p^r)^b
